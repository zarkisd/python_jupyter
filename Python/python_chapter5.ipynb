# sum_all_basic.py

```python
def sum_all(start,end):
    output=0
    for i in range(start, end+1):
        output+=i
    return output

print(f"0 to 100 : {sum_all(0,100)}")
print(f"0 to 1000 : {sum_all(0,1000)}")
print(f"50 to 100 : {sum_all(50,100)}")
print(f"500 to 1000 : {sum_all(500,1000)}")
```

# sum_all_with_default.py

```python
def sum_all(start=0, end=100, step=1):
    output=0
    for i in range(start, end+1, step):
        output+=i
    return output

print(f"A. {sum_all(0,100,10)}")
print(f"B. {sum_all(end=100)}")
print(f"C. {sum_all(end=100, step=2)}")
```

# factorial_for.py

```python
def factorial(n):
    output=1
    for i in range(1,n+1):
        output*=i
    return output
print(f"1! : {factorial(1)}")
print(f"2! : {factorial(2)}")
print(f"3! : {factorial(3)}")
print(f"4! : {factorial(4)}")
print(f"5! : {factorial(5)}")
print()
def factorial1(n):
    if n==1:
        return 1
    else:
        return n*factorial1(n-1)
print(f"1! : {factorial1(1)}")
print(f"2! : {factorial1(2)}")
print(f"3! : {factorial1(3)}")
print(f"4! : {factorial1(4)}")
print(f"5! : {factorial1(5)}")
```

# fibonacci_recursion01.py

```python
counter=0
def fibonacci(n):
    
    global counter
    counter+=1
    if n==1:
        return 1
    if n==2:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
print(f"fibonacci(1) : {fibonacci(1)}, 덧셈 횟수 : {counter}")
print(f"fibonacci(2) : {fibonacci(2)}, 덧셈 횟수 : {counter}")
print(f"fibonacci(3) : {fibonacci(3)}, 덧셈 횟수 : {counter}")
print(f"fibonacci(4) : {fibonacci(4)}, 덧셈 횟수 : {counter}")
print(f"fibonacci(5) : {fibonacci(5)}, 덧셈 횟수 : {counter}")
```

```python
#2
#생각방법
# 1번째 테이블 앉을사람 수 결정
# 남은 사람은 남은테이블 배치
# 1번째 테이블 앉을사람 결정하면 테이블 수 1개 줄어들고 남은 사람수는 전체에서 1번쨰 테입르 사람수 빼면 알수있다
# 테이블수는 제한 없으므로 사람 수를 나누는 패턴만 생각
# 따라서 이전 테이블에 배치한 인원수와 같거나 많게 반복할당
# 남은 사람수와 이전 테이블에 배치한 사람수만 알면 재귀함수 가능

# 브루트 포스방법이라고 함 전체탐색

#핸드폰 메모에 이거 스케치 있음

#앉힐수있는최소사람수
m=2
#앉힐수있는최대사람수
M=10
#전체사람의수
P=100
memo={}
# 메모화 시키기 위한 변수
number=0

def problem(leave, sit):
    key=str([leave, sit])
    
    #종료조건
    if key in memo:
        return memo[key]
    if leave <0:
        #무효이므로 0 리턴
        return 0
    if leave ==0:
        #유효하므로 수를 세기 위해 1을 리턴
        return 1
    #재귀처리
    cnt=0
    for i in range(sit, M+1):
        cnt+=problem(leave-i, i)
    
    #메모화 처리
    memo[key]=cnt
    global number
    number+=1
    #종료
    return cnt
    
print(problem(P, m))
print("실행횟수 : ",number)
```

```python
import random
hanguls=list("가나라다라마바사아자카타파하")
with open("info.txt","w") as file:
    for i in range(1000):
        name=random.choice(hanguls)+random.choice(hanguls)
        weight=random.randrange(40,100)
        height=random.randrange(140,200)
        file.write(f"{name}, {weight}, {height}\n")
```

```python
with open("info.txt","r") as file:
    for line in file:
        (name, weight, height)=line.strip().split(", ")
        
        if (not name) or (not weight) or(not height):
            continue
        bmi=int(weight)/((int(height)/100)**2)
        result=""
        if 25<=bmi:
            result="과체중"
        elif 18.5<=bmi:
            result="정상 체중"
        else:
            result="저체중"
            
        print("\n".join([
         "이름 : {}",
        "몸무게 : {}",
            "키 : {}",
            "BMI : {}",
            "결과 : {}"
            
        ]       
        ).format(name, weight, height, bmi, result))
        print()
            
```

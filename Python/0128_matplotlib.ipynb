# 교통사고가 가장 많이 발생하는 시간대 알아보기
- "발생년월일시'

```python
time=df["발생년월일시"]%100
```

```python
# copy는 데이터만 가져가서 df 원본 데이터에 영향 X
minute=df["발생분"].copy()
```

```python
# 0~2
# 3~5
# 6~8
# 9~11
# 12~14
# 15~17
# 18~20
# 21~23
# 카테고리 만들기
```

```python
bins=[-1,2,5,8,11,14,17,20,23]
labels=["0~2","3~5", "6~8", "9~11", "12~14", "15~17", "18~20","21~23"]
z=pd.cut(time,bins,labels=labels)
```

```python
# 시간별로 몇건의 사고가 발생했는지 세보기
print(z.value_counts())
# 시간순으로 정렬
print(z.value_counts().sort_index())
rs=z.value_counts().sort_index()
```

# Pie차트 그리기

```python
plt.figure(figsize=(5,5))
#labels 각 데이터에 이름 달아줌
# autopct 각 데이터가 몇 퍼센트를 차지하고 있는지 알려줌
plt.pie(rs,labels=labels,autopct="%.2f%%")
plt.show()
```

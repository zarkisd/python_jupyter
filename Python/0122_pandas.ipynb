```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])
```

```python
display(df1,df2,df3)
```

```python
#concat 활용 , 기본 방향 위에서 아래로
pd.concat([df1,df2,df3])
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'], 
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
display(df4)
```

```python
pd.concat([df1,df2,df3,df4])
# axis=1
# index방향으로 데이터를 합침
pd.concat([df1,df2,df3,df4], axis=1)
```

```python
# join 속성
# outer, inner
display(pd.concat([df1,df4], axis=1, join="outer"))
# 전체 데이터 출력, 기본값
print("************")
print()
display(pd.concat([df1,df4], axis=1, join="inner"))
# 공유하고있는 기준값만 출력

#원하는 인덱스의 값만 가져오기
# reindex함수 사용
display(pd.concat([df1,df4],axis=1).reindex(df1.index))
```

```python
#인덱스(기준이 아닌거) 순서 재졍렬 선생님말씀
# 기준 인덱스를 무시하고 새로운 인덱스를 부여
# ignore_index
display(pd.concat([df1,df4]))
print()
display(pd.concat([df1,df4], ignore_index=True))
print()
display(pd.concat([df1,df4], ignore_index=True, axis=1))
```

# 컬럼 만들고 새로운 값 채우기
- apply 함수
- 기준컬럼 생성
- 값에는 단가 값을 비교해서 상, 중, 하의 데이터를 기준 컬럼에 채움

```python
df["단가"].describe()
```

```python
# 상 500 이상
# 중 500미만 226이상
# 하 226 미만
```

```python
def make_standard(apt):
    if apt["단가"]>500:
        return "상"
    elif apt["단가"]>=226:
        return "중"
    else:
        return "하"
```

```python
df["기준"]=df.apply(make_standard, axis=1)
```

```python
display(df)
```

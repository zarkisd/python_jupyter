```python
data=np.loadtxt("height_weight.txt", delimiter=",")
print(data)
```

```python

list=[]
for i in range(0,10):
    list1=(data[1][i])/((data[0][i]/100)*(data[0][i]/100))
    list.append(list1)
print(list)

for i in range(0,10):
    print("%1.2f"%list[i])
print()
for i in range(0,10):
    print("{:1.3f}".format(list[i]))
```

```python
data=np.loadtxt("ratings.txt", delimiter="::")
data=data.astype("int32")
print(data)
```

```python
data.shape
```

```python
#전체 평점 평균 구하기
print(np.mean(data[:,2]))
```

```python
# 1번 사용자의 평점 평균 구하기
# boolean 인덱싱
#print(np.mean(data[num==1,:],2))
bol = np.array([True , False, True, False])
num = np.array(data[ : , 0])
score = np.array(data[ : , 2])
print(num)
print(score)
m=np.mean(score[num == 1])
print("{0:1.3f}".format(m))
```

```python
data[data[:,0]==1,2].mean()
```

```python
# 각 사용자별 평균 평점
# unique
# 중복된 값들 한번만 출력
print(data[:,0])
print(np.unique(data[:,0]))
```

```python
user_id=np.unique(data[:,0])
for userid in user_id:
    print(userid)
```

```python
user_id.size
```

```python
user_id=np.unique(data[:,0])
for userid in user_id:
    #print(userid, "번 사용자",data[data[:,0]==userid,2].mean() )
    print(f"{userid}번 사용자의 평균 평점 : %.2f"%data[data[:,0]==userid,2].mean())
```
